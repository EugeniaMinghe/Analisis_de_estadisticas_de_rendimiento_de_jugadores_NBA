###### GAME

```python
game= 'import\game.csv'

games = pd.read_csv(game)

games.head()
```

```python
games.info()
```

###### Convertir game_date a datetime

```python
games['game_date']= pd.to_datetime(games['game_date'])
```

```python
games.info()
```

###### Filtramos por nombre del equipo

```python
los_angeles_lakers = ['LAL']
juegos_filtrados = games[
(games['team_abbreviation_home'].isin(los_angeles_lakers))| 
(games['team_abbreviation_away'].isin(los_angeles_lakers))
]

juegos_filtrados.info()
```

###### Filtrar por la fecha del juego para tener las últimas tres temporadas que me interesan

```python
ultimas_tres = juegos_filtrados[juegos_filtrados['game_date'] >= '2020-10-01']
ultimas_tres.head(20)
```

```python
ultimas_tres.to_csv('export\games_lakers_tres_temporadas.csv', index=False)
```

###### LINE_SCORE

###### Filtrar este csv por los game_id de interés, es decir, los guardados en games_lakers_tres_temporadas

```python

# Cargar ambos CSVs
line_score = pd.read_csv("import\line_score.csv")
game = pd.read_csv("export\games_lakers_tres_temporadas.csv")

# Crear una máscara con los game_id que están en el games_lakers_tres_temporadas
mascara = line_score['game_id'].isin(game['game_id'])

# Aplicar la máscara para filtrar el DataFrame
line_score_filtered = line_score[mascara]

print(line_score_filtered)
```
