### Matriz de correlación de variables numéricas

```python
# Creo una matriz de correlación de todas las variables numéricas de interés
numeric_columns = df_games_lakers.select_dtypes(include = (['float64'])).columns
correlation_matrix = df_games_lakers[numeric_columns].corr()

# Creo una máscara para que no se muestre el triángulo superior de la matriz de correlación y simplificar su interpretación
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool)) 

# La siguiente línea de código es para mostrar solamente las correlaciones fuertes (mayor a 0,5) ya sean directas o inversas.
correlation_matrix = correlation_matrix[abs(correlation_matrix) > abs(0.5)]

# Creo la gráfica
plt.figure(figsize = (25,10))

# Creo el heatmap
sns.heatmap(correlation_matrix, 
            mask=mask, 
            annot=True, 
            cmap='coolwarm', 
            fmt='.2f', 
            linewidths=0.5,      # Grosor de las líneas de cuadrícula
            linecolor='black',  # Color de las líneas de cuadrícula
            vmin=-1, vmax=1,  
            cbar_kws={'shrink': 0.8})

plt.title('Heatmap de Correlación - Lakers', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.show()
```

#### Gráfico de barras agrupadas para mostrar la efectividad promedio en cada tipo de tiro (de local y de visitante)

```python
# Filtrar solo los partidos donde los Lakers fueron locales o visitantes
df_lakers = df_games_lakers[
    (df_games_lakers["team_name_home"] == "Los Angeles Lakers") | 
    (df_games_lakers["team_name_away"] == "Los Angeles Lakers")
].copy()

# Asignar la efectividad correcta según la condición de local o visitante
df_lakers["FG% Local"] = df_lakers["fg_pct_home"].where(df_lakers["team_name_home"] == "Los Angeles Lakers")
df_lakers["FG% Visitante"] = df_lakers["fg_pct_away"].where(df_lakers["team_name_away"] == "Los Angeles Lakers")

df_lakers["3PT% Local"] = df_lakers["fg3_pct_home"].where(df_lakers["team_name_home"] == "Los Angeles Lakers")
df_lakers["3PT% Visitante"] = df_lakers["fg3_pct_away"].where(df_lakers["team_name_away"] == "Los Angeles Lakers")

df_lakers["FT% Local"] = df_lakers["ft_pct_home"].where(df_lakers["team_name_home"] == "Los Angeles Lakers")
df_lakers["FT% Visitante"] = df_lakers["ft_pct_away"].where(df_lakers["team_name_away"] == "Los Angeles Lakers")

# Calcular el promedio de efectividad de los Lakers en cada tipo de tiro y armar DataFrame
df_efectividad = pd.DataFrame({
    "Tipo de Tiro": ["Tiros de Campo", "Tiros de Triple", "Tiros Libres"],
    "Local": [df_lakers["FG% Local"].mean(), df_lakers["3PT% Local"].mean(), df_lakers["FT% Local"].mean()],
    "Visitante": [df_lakers["FG% Visitante"].mean(), df_lakers["3PT% Visitante"].mean(), df_lakers["FT% Visitante"].mean()]
})

# Transformar los datos para seaborn
df_efectividad_melted = df_efectividad.melt(id_vars="Tipo de Tiro", var_name="Condición", value_name="Efectividad")

# Crear gráfico de barras agrupadas
plt.figure(figsize=(8,6))
sns.barplot(data=df_efectividad_melted, x="Tipo de Tiro", y="Efectividad", hue="Condición", palette="deep")

# Personalización del gráfico
plt.xlabel("")
plt.ylabel("Porcentaje de Efectividad")
plt.title("Comparación de Efectividad de los Lakers por Tipo de Tiro (Local vs Visitante)")
plt.ylim(0, 1)  # Ya que los porcentajes están en decimal (0-1)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostrar el gráfico
plt.show()
```

# Análisis por temporada

 Crear la columna 'season' teniendo en cuenta que si game_date es entre enero y junio, la temporada a la que corresponde el partido es el año anterior que game_date y si es entre junio y diciembre la temporada es el mismo año que game_date (una temporada de NBA va de octubre de un año a junio del siguiente).

```python
df_games_lakers['game_date'] = pd.to_datetime(df_games_lakers['game_date'])
df_games_lakers['season'] = df_games_lakers['game_date'].apply(
    lambda x: f"{x.year}-{x.year + 1}" if x.month >= 7 else f"{x.year - 1}-{x.year}")
```

```python
 # Obtener la última temporada disponible
ultima_temporada = df_games_lakers['season'].max()

# Extraer los tres últimos años de la temporada (ej: de "2021-2022" a 2023)
ultima_temporada_inicio = int(ultima_temporada.split('-')[0])

# Generar una lista con las últimas 3 temporadas en formato "YYYY-YYYY"
ultimas_3_temporadas = [f"{año}-{año+1}" for año in range(ultima_temporada_inicio-2, ultima_temporada_inicio+1)]


# Verificar los datos
print(df_games_lakers['season'].unique()) 

```

```python
# Contar la cantidad de partidos jugados en cada temporada
partidos_por_temporada = df_games_lakers.groupby('season')['game_date'].count().reset_index()

# Renombrar columnas para mayor claridad
partidos_por_temporada.columns = ['Temporada', 'Partidos Jugados']

# Mostrar los resultados
print(partidos_por_temporada)
```

### Gráfico de líneas múltiples para mostrar los puntos totales, los puntos de local y los puntos de visitante anotados por Los Lakers en cada temporada.

```python
# Filtrar partidos donde los Lakers fueron locales y sumar puntos por temporada
puntos_local = df_games_lakers[df_games_lakers['team_name_home'] == 'Los Angeles Lakers'] \
                .groupby('season')['pts_home'].sum().reset_index()
puntos_local.rename(columns={'pts_home': 'points'}, inplace=True)
puntos_local['tipo'] = 'Local'

# Filtrar partidos donde los Lakers fueron visitantes y sumar puntos por temporada
puntos_visitante = df_games_lakers[df_games_lakers['team_name_away'] == 'Los Angeles Lakers'] \
                    .groupby('season')['pts_away'].sum().reset_index()
puntos_visitante.rename(columns={'pts_away': 'points'}, inplace=True)
puntos_visitante['tipo'] = 'Visitante'

# Sumar los puntos totales de cada temporada (local + visitante)
puntos_totales = df_games_lakers[df_games_lakers['team_name_home'] == 'Los Angeles Lakers'] \
                   .groupby('season')['pts_home'].sum() + \
                 df_games_lakers[df_games_lakers['team_name_away'] == 'Los Angeles Lakers'] \
                   .groupby('season')['pts_away'].sum()

puntos_totales = puntos_totales.reset_index()
puntos_totales.rename(columns={0: 'points'}, inplace=True)
puntos_totales['tipo'] = 'Total'

# Unimos los tres DataFrames
df_puntos = pd.concat([puntos_local, puntos_visitante, puntos_totales])


plt.figure(figsize=(10,6))

# Graficar líneas de puntos totales, local y visitante
sns.lineplot(data=df_puntos, x='season', y='points', hue='tipo', marker='o', linewidth=2.5)

# Ajustes del gráfico
plt.title('Puntos Totales Anotados por LAL por Temporada (Total, Local y Visitante)', fontsize=14)
plt.xlabel('Temporada', fontsize=12)
plt.ylabel('Puntos Totales', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Tipo de Partidos")

plt.show()
```

### El mismo gráfico de líneas que arriba pero solo comparando los puntos como visitante y como local (para apreciar mejor la dif. entre ambos)

```python
df_puntos_local_visitante = df_puntos[df_puntos['tipo'].isin(['Local', 'Visitante'])]

# Graficar solo local y visitante
plt.figure(figsize=(10,6))
sns.lineplot(data=df_puntos_local_visitante, x='season', y='points', hue='tipo', marker='o', linewidth=2.5)

# Ajustes del gráfico
plt.title('Puntos Totales Anotados por LAL por Temporada (Local y Visitante)', fontsize=14)
plt.xlabel('Temporada', fontsize=12)
plt.ylabel('Puntos Totales', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Condición")

plt.show()
```

### Gráfico de barras agrupadas que muestra la cantidad de partidos ganados por LAL en cada temporada

```python
# Filtrar partidos ganados por los Lakers
df_games_lakers['win_home'] = (df_games_lakers['team_name_home'] == 'Los Angeles Lakers') & (df_games_lakers['wl_home'] == 'W')
df_games_lakers['win_away'] = (df_games_lakers['team_name_away'] == 'Los Angeles Lakers') & (df_games_lakers['wl_away'] == 'W')

# Contar victorias totales, locales y visitantes por temporada
wins_per_season = df_games_lakers.groupby('season').agg(
    home_wins=('win_home', 'sum'),
    away_wins=('win_away', 'sum')
).reset_index()

# Calcular el total de victorias como la suma de las victorias locales y de visitante
wins_per_season['total_wins'] = wins_per_season['home_wins'] + wins_per_season['away_wins']

# Gráfico de barras agrupadas
plt.figure(figsize=(10, 6))

bar_width = 0.25
seasons = wins_per_season['season']

# Posiciones de las barras
x = range(len(seasons))

# Graficar las barras
home_bars = plt.bar([i - bar_width for i in x], wins_per_season['home_wins'], width=bar_width, label="Ganados Local", color='blue', alpha=0.7)
away_bars = plt.bar(x, wins_per_season['away_wins'], width=bar_width, label="Ganados Visitante", color='red', alpha=0.7)
total_bars = plt.bar([i + bar_width for i in x], wins_per_season['total_wins'], width=bar_width, label="Total Ganados", color='purple', alpha=0.7)

# Agregar números sobre las barras
for bars in [home_bars, away_bars, total_bars]:
    for bar in bars:
        height = bar.get_height()
        # Colocar el número sobre la barra
        plt.text(bar.get_x() + bar.get_width() / 2, height + 0.2, str(int(height)), ha='center', va='bottom', fontsize=10)

# Personalizar gráfico
plt.xlabel("Temporada")
plt.ylabel("Cantidad de Partidos Ganados")
plt.title("Cantidad de Partidos Ganados por Temporada (Lakers)")
plt.xticks(x, seasons, rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Mostrar gráfico
plt.show()
```

En el gráfico de arriba vemos que en la temporada 2020-2021 se ganaron más partidos que en la 2021-2022, además el rendimiento de local y visitante fue mucho más parejo.

### Gráfico de barras agrupadas que muestra los puntos totales anotados por LAL en cada temporada

```python
# Sumar los puntos anotados por los Lakers como local y como visitante
df_games_lakers['pts_home_lakers'] = df_games_lakers['pts_home'] * (df_games_lakers['team_name_home'] == 'Los Angeles Lakers')
df_games_lakers['pts_away_lakers'] = df_games_lakers['pts_away'] * (df_games_lakers['team_name_away'] == 'Los Angeles Lakers')

# Sumar puntos anotados como local y como visitante por temporada
points_per_season = df_games_lakers.groupby('season').agg(
    home_points=('pts_home_lakers', 'sum'),
    away_points=('pts_away_lakers', 'sum')
).reset_index()

# Calcular el total de puntos como la suma de los puntos como local y de visitante
points_per_season['total_points'] = points_per_season['home_points'] + points_per_season['away_points']

# Gráfico de barras agrupadas
plt.figure(figsize=(10, 6))

bar_width = 0.25
seasons = points_per_season['season']

# Posiciones de las barras
x = range(len(seasons))

# Graficar las barras
home_bars = plt.bar([i - bar_width for i in x], points_per_season['home_points'], width=bar_width, label="Puntos Local", color='blue', alpha=0.7)
away_bars = plt.bar(x, points_per_season['away_points'], width=bar_width, label="Puntos Visitante", color='red', alpha=0.7)
total_bars = plt.bar([i + bar_width for i in x], points_per_season['total_points'], width=bar_width, label="Total Puntos", color='purple', alpha=0.7)

# Agregar números sobre las barras
for bars in [home_bars, away_bars, total_bars]:
    for bar in bars:
        height = bar.get_height()
        # Colocar el número sobre la barra
        plt.text(bar.get_x() + bar.get_width() / 2, height + 50, str(int(height)), ha='center', va='bottom', fontsize=10)

# Personalizar gráfico
plt.xlabel("Temporada")
plt.ylabel("Puntos Anotados")
plt.title("Puntos Anotados por Temporada (Lakers)")
plt.xticks(x, seasons, rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Mostrar gráfico
plt.show()
```

#### Si comparamos los gráficos de partidos ganados y puntos anotados se evidencia la vital importancia de un buen rendimiento en la faz defensiva (incluso más que la ofensiva) para la obtención de victorias, ya que se anotaron más puntos en la temporada 21-22 que en la 20-21, sin embargo en la temprada 20-21 LAL ganaron muchos más partidos (un 50% más) que en la temporada 21-22.

#### Comparación efectividad en tiros de campo y tiros de triple entre temporadas.

```python
# Filtrar los datos para incluir solo los partidos de los Lakers
df_shooting_lakers = df_games_lakers[(df_games_lakers['team_name_home'] == 'Los Angeles Lakers') | 
                                     (df_games_lakers['team_name_away'] == 'Los Angeles Lakers')]

# Filtrar por las temporadas deseadas
df_shooting_lakers = df_shooting_lakers[df_shooting_lakers['season'].isin(['2020-2021', '2021-2022', '2022-2023'])]

# Calcular FG% promedio por temporada y condición de local/visitante
fg_data = df_shooting_lakers.groupby('season')[['fg_pct_home', 'fg_pct_away']].mean().reset_index()
fg_data = fg_data.melt(id_vars='season', var_name='Local/Visitante', value_name='FG%')

# Calcular 3PT% promedio por temporada y condición de local/visitante
three_pt_data = df_shooting_lakers.groupby('season')[['fg3_pct_home', 'fg3_pct_away']].mean().reset_index()
three_pt_data = three_pt_data.melt(id_vars='season', var_name='Local/Visitante', value_name='3PT%')

# Graficar FG%
plt.figure(figsize=(8,5))
sns.barplot(data=fg_data, x='season', y='FG%', hue='Local/Visitante', palette=['#1f77b4', '#ff7f0e'])
plt.title('Comparación de FG% entre Temporadas (Lakers)')
plt.ylabel('FG% Promedio')
plt.show()

# Graficar 3PT%
plt.figure(figsize=(8,5))
sns.barplot(data=three_pt_data, x='season', y='3PT%', hue='Local/Visitante', palette=['#1f77b4', '#ff7f0e'])
plt.title('Comparación de 3PT% entre Temporadas (Lakers)')
plt.ylabel('3PT% Promedio')
plt.show()
```

```python
# Filtrar los datos para incluir solo los partidos de los Lakers
df_lakers = df_shooting[(df_shooting['team_name_home'] == 'Los Angeles Lakers') | 
                        (df_shooting['team_name_away'] == 'Los Angeles Lakers')]

# Calcular promedios de estadísticas defensivas
dreb_data = df_lakers.groupby('season')[['dreb_home', 'dreb_away']].mean().reset_index()
stl_data = df_lakers.groupby('season')[['stl_home', 'stl_away']].mean().reset_index()
blk_data = df_lakers.groupby('season')[['blk_home', 'blk_away']].mean().reset_index()

# Transformar los datos para gráficos
dreb_data = dreb_data.melt(id_vars='season', var_name='Local/Visitante', value_name='Rebotes Defensivos')
stl_data = stl_data.melt(id_vars='season', var_name='Local/Visitante', value_name='Robos')
blk_data = blk_data.melt(id_vars='season', var_name='Local/Visitante', value_name='Bloqueos')

# Graficar Rebotes Defensivos
plt.figure(figsize=(8,5))
sns.barplot(data=dreb_data, x='season', y='Rebotes Defensivos', hue='Local/Visitante', palette=['#1f77b4', '#ff7f0e'])
plt.title('Comparación de Rebotes Defensivos entre Temporadas (Lakers)')
plt.ylabel('Promedio de Rebotes Defensivos')
plt.xlabel('Temporada')
plt.legend(title='Condición')
plt.show()

# Graficar Robos
plt.figure(figsize=(8,5))
sns.barplot(data=stl_data, x='season', y='Robos', hue='Local/Visitante', palette=['#1f77b4', '#ff7f0e'])
plt.title('Comparación de Robos entre Temporadas (Lakers)')
plt.ylabel('Promedio de Robos')
plt.xlabel('Temporada')
plt.legend(title='Condición')
plt.show()

# Graficar Bloqueos
plt.figure(figsize=(8,5))
sns.barplot(data=blk_data, x='season', y='Bloqueos', hue='Local/Visitante', palette=['#1f77b4', '#ff7f0e'])
plt.title('Comparación de Bloqueos entre Temporadas (Lakers)')
plt.ylabel('Promedio de Bloqueos')
plt.xlabel('Temporada')
plt.legend(title='Condición')
plt.show()
```

Estos gráficos de estadísticas defensivas reafirman la gran importancia de un buen desempeño en la faz defensiva. Se observa un promedio más alto de rebotes defensivos y bloqueos en las temporadas con más partidos ganados.
